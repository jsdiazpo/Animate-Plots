```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.facecolor'] = 'w'
```

```python
from celluloid import Camera
```

```python
fig = plt.figure()
camera = Camera(fig)
x = np.linspace(0, 100)
for i in range(20):
    t = plt.plot(x, np.cos(x+i/20))
    plt.legend(t, [f'line {i}'])
    camera.snap()
animation = camera.animate()
animation.save('celluloid_legends.gif', writer = 'imagemagick')
```

```python
from matplotlib.animation import FuncAnimation
from IPython import display
```

```python
x = np.linspace(0, 2*np.pi, 100)

fig, ax = plt.subplots()
lines = ax.plot([])
line = lines[0]

# other static stuff
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-2, 2)

def animate(frame):
    # update plot
    y = np.sin(x + 2*np.pi * frame/100)
    line.set_data((x, y))
    
animt = FuncAnimation(fig, animate, frames=100, interval=20)
video = animt.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```

```python

fig, ax = plt.subplots()
lines = ax.plot([], 'o')
line = lines[0]

# other static stuff
ax.set_aspect('equal')
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

def animate(frame):
    # update plot
    x = np.cos(2*np.pi * frame/100)
    y = np.sin(2*np.pi * frame/100)
    line.set_data((x, y))

num_loops = 2
    
animt = FuncAnimation(fig, animate, frames=100*num_loops, interval=30, repeat=True)
video = animt.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```

```python
animt.save('animt.gif', writer='imagemagick')
```
